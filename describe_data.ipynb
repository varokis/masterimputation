```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


sns.set_theme()
sns.set_palette('deep')
```

# Visualize and describe the data

3 data sets with sampling rates of 2,4,6h and 5,5,6 dynamic features respectively

```python
num_cols = ['heart rate', 'systolic blood pressure', 'diastolic blood pressure', 'respiratory rate', 'oxygen saturation'] # + 'temperature' for the 6h dataset

# load the data
df2h = pd.read_parquet('outdata/datasets/complete/data2h.parquet', columns=num_cols)
df4h = pd.read_parquet('outdata/datasets/complete/data4h.parquet', columns=num_cols)
df6h = pd.read_parquet('outdata/datasets/complete/data6h.parquet', columns=num_cols + ['temperature'])
```

## 2h sampling

```python
df2h.describe().round(2)
```

```python
# 2h
fig = plt.figure(constrained_layout=True, figsize=(10, 15))
fig.suptitle('2h Sampling - Feature Distribution', fontsize=16)
subfigs = fig.subfigures(3,1)

# heart rate and respiratory rate 
hr_rr = subfigs[0].subplots(1,2)
sns.histplot(data=df2h, x='heart rate', ax=hr_rr[0], bins = 50, kde=True)
hr_rr[0].set_title('Heart rate')
hr_rr[0].set_xlabel('Heart rate [bpm]')
# # vertical lines for max, min and median
# for q in df2h['heart rate'].quantile([0, 0.25, 0.5, 0.75, 1]):
#     hr_rr[0].axvline(q, color='red', linestyle='--')


sns.histplot(data=df2h, x='respiratory rate', ax=hr_rr[1], bins = 50, kde=True)
hr_rr[1].set_title('Respiratory rate')
hr_rr[1].set_xlabel('Respiratory rate [bpm]')

# blood pressure
bp = subfigs[1].subplots(1,2)
sns.histplot(data=df2h, x='systolic blood pressure', ax=bp[0], bins = 50, kde=True)
bp[0].set_title('Systolic blood pressure')
bp[0].set_xlabel('Systolic blood pressure [mmHg]')
sns.histplot(data=df2h, x='diastolic blood pressure', ax=bp[1], bins = 50, kde=True)
bp[1].set_title('Diastolic blood pressure')
bp[1].set_xlabel('Diastolic blood pressure [mmHg]')

# spO2
spo2 = subfigs[2].subplots(1,2)
sns.histplot(data=df2h, x='oxygen saturation', ax=spo2[0], bins = 50, kde=True)
spo2[0].set_title('Oxygen saturation')
spo2[0].set_xlabel('Oxygen saturation [%]')
spo2[1].axis('off')

plt.show()
```
